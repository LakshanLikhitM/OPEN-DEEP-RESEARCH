```python
# Simple Chatbot for 3 Specific Questions

print("Chatbot: Hello! Ask me something. Type 'bye' to exit.\n")

while True:
    user_input = input("You: ").lower()

    if user_input == "bye":
        print("Chatbot: Goodbye!")
        break

    elif user_input == "hi":
        print("Chatbot: Hello! Nice to meet you!")

    elif user_input == "how are you":
        print("Chatbot: I'm doing great, thank you!")

    elif user_input == "is your day great":
        print("Chatbot: Yes! My day is great. Thanks for asking!")

    else:
        print("Chatbot: Sorry, I can only answer: hi, how are you, or is your day great.")
```
